## A. Historical stock trends using interactive candlestick charts

```python
#Callback decorator
@app.callback(Output('Stock-graph', 'figure'), [Input('stock-list', 'value')])

def graph(selected_dropdown_value):
    
    aggs = cast(
        HTTPResponse,
        client.get_aggs(
            selected_dropdown_value,
            1,
            'day',
            '2022-08-29',
            '2023-08-29',
            adjusted=True,
            raw=True,
            limit=5000
             
        ),
        
    )
    data = json.loads(aggs.data)
    
    for item in data:
        if item == 'results':
            rawData = data[item]
    
    closelist = []
    openlist = []
    highlist = []
    lowlist = []
    #volumelist = []#too much data for candlestick
    timelist = []
    
    for bar in rawData:
        for category in bar:
            if category == 'c':
                closelist.append(bar[category])
            elif category == 'o':
                openlist.append(bar[category])
            elif category == 'h':
                highlist.append(bar[category])
            elif category == 'l':
                lowlist.append(bar[category])
            elif category == 't':
                timelist.append(pd.Timestamp(bar[category], tz='GMT', unit='ms'))#to change GMT and milisecond formats
                                             
    fig = go.Figure()
    fig.add_trace(go.Candlestick(x=timelist, close=closelist, open=openlist, high=highlist, low=lowlist, name='Stock Data'))
    fig.update_layout(xaxis_rangeslider_visible=False, template='plotly_dark') #to eliminate slider plotly outputs by default and use dark mode 
    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8050)

#Alternative access to charts by pasting this link, http://127.0.0.1:8050/ , on web browser
```

## Calculating the Exponential Moving Average (EMA) indicator

```python
import talib
import numpy as np
import datetime as dt
```

```python
def getData():
     client = rest.RESTClient(config.API_KEY)
     closelist = []
     timelist = []
     #endDate = dt.datetime.today().date()#can be used for dynamic time/date updating (from 2022-08-23 onwards)
     
     aggs = cast(
        HTTPResponse,
        client.get_aggs(
        varname,
        1,
        'day',
        '2022-08-29',
        '2023-08-29',
        adjusted=True, 
        raw=True,
        limit=5000
            ),
    )
     data = json.loads(aggs.data)
     bars = data['results']
     
     for bar in bars:
         for value in bar:
             if value == 'c':
                 closelist.append(bar[value])
             elif value == 't':
                 timelist.append(pd.Timestamp(bar[value], tz='GMT', unit='ms'))

     #convert closelist into a numpy array for data processing using talib
     closelist = np.array(closelist)
     #get talib EMAs
     ema_eight = talib.EMA(closelist, 8)
     ema_thirteen = talib.EMA(closelist, 13)
     ema_twentyone = talib.EMA(closelist, 21)
     ema_fiftyfive = talib.EMA(closelist, 55)
     
     print(f'\n\033[1mEMA_50 calculations for {varname} stocks:\n\033[0m')
     
     # checking for date when EMA_55 crosses above or under the other EMAs,
     # which marks the entry/buy and the exit/sell points, respectively.
     for i in range(len(ema_fiftyfive)):
         if (ema_fiftyfive[i] > ema_eight[i]) and (ema_fiftyfive[i] > ema_thirteen[i]) and (ema_fiftyfive[i] > ema_twentyone[i]):
             print("EMA_50 has crossed ABOVE ALL other EMAs on: " + str(timelist[i]) )
         elif(ema_fiftyfive[i] < ema_eight[i]) and (ema_fiftyfive[i] < ema_thirteen[i]) and (ema_fiftyfive[i] < ema_twentyone[i]):
             print("EMA_50 has crossed UNDER ALL other EMAs on: " + str(timelist[i]) )
getData()
```

## Calculating the Simple Moving Average (SMA) and the Relative Strength Index (RSI) indicators

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
import warnings
warnings.filterwarnings('ignore')
```

```python
client = rest.RESTClient(config.API_KEY)

aggs = cast(
    HTTPResponse,
    client.get_aggs(
        varname,
        1,
        'day',
        '2022-08-29',
        '2023-08-29',
        adjusted=True,
        raw=True,
        limit=5000
    )
)

poly_data = json.loads(aggs.data)
poly_data = poly_data['results'] #to get the bars data only (e.g., o, h, l, etc.)

dates = []  
for bar in poly_data:
    for item in bar:
        if item == 't':
            dates.append(pd.Timestamp(bar[item], tz='GMT', unit='ms').date()) 

df = pd.DataFrame(poly_data)

d = {'Open' : df['o'], 'High' : df['h'], 'Low' : df['l'], 'Close': df['c'], 'Volume': df['v']} #get only bars needed for backtesting

dataFrame = pd.DataFrame(d)
pd.DateTimeIndex = dates #displays graph but no dates
#dataFrame.index = dates #first codeline tried but not working


#uncomment lines below to pirnt and compare df's
# print(df)
# print(dataFrame)

#Backtesting analysis SMA and RSI

class smacross(Strategy):
    def init(self):
        #n1 = 10
        #n2 = 20
        price = self.data.Close
        self.sma1 = self.I(SMA, price, n1)
        self.sma2 = self.I(SMA, price, n2)
    
    def next(self):
        if crossover(self.sma1, self.sma2):
            self.buy()
        elif crossover(self.sma2, self.sma1):
            self.sell()
bt = Backtest(dataFrame, smacross, commission =0.002, exclusive_orders=True) #exclusive_orders is an optional arg

print(f'\n\033[1mSMA Backtest Output on {varname} Stocks (Stats & Interactive Charts):\n\033[0m.')
print(bt.run())
print()
stats = bt.run()
bt.plot()



class rsioscillator(Strategy): #using typical values as described in talib and backtesting github page 
    upper_bound = upper_bound
    lower_bound = lower_bound
    rsi_window = rsi_window

    def init(self):
        self.rsi = self.I(talib.RSI, self.data.Close, self.rsi_window)

    def next(self):
        if crossover(self.rsi, self.upper_bound):
            self.position.close()
        elif crossover(self.lower_bound, self.rsi):
            self.buy()
bt = Backtest(dataFrame, rsioscillator, cash=10000, commission=.002)

print(f'\n\033[1mRSI Backtest Output on {varname} Stocks (Stats & Interactive Charts):\n\033[0m.')
print(bt.run())
print()
stats = bt.run()
bt.plot()
```

## Understanding and Interpreting Candlestick Charts:
Candlestick charts was created by the Joriginated in Japan who discovered that in addition to the influence of supply and demand on the price of rice, the the markets were also strongly based on the expectations and emotions (fear and greed) of traders **[1]** **[2]**. Candlestick charts allow to show that emotion by visually representing the size of price moves with different colors. Traders use the candlesticks to make trading decisions based on irregularly occurring patterns that help forecast the short-term direction of the price. Candlestick trend patterns portray trader sentiment over trading periods. There is no "most accurate" pattern as they should all be viewed as indicators of what bull or bear traders might be thinking —but some traders have preferences and act on specific patterns— **[2]**.

## Understanding and Interpreting Backtesting Strategy Results:
In our Python tool, we compute the EMS, SMA and RSI indicators for implemeting a backtesting strategy. Like in the case of candlestick charts, a dataset must contain OHLC values for each time period considered when computating these three indicators. Values for volume may also be added to the dataset, which are then included in the candlesticks charts returned in the respective backtesting results.

Backtesting is the general method for seeing how well a strategy or a predictive model would have done after the fact **[3]**. In other words, backtesting is method that helps traders and investors simulate their trading strategies on past (historical) data to gauge their effectiveness and potential profitability before using them in real-time trading. When traders are backtesting a day trading strategy (15-minute timeframe or lower), it is usually enough to go back two to three months and start the backtest there. When traders are backtesting a strategy on a higher timeframe, they will have to go back 6 to 12 months **[4]**. 

In backtesting, the EMA gives more weight to the most recent prices, thereby aligning the average closer to current prices. In general, short-term traders typically rely on the 12- or 26-day EMA, while the ever-popular 50-day and 200-day EMA is used by long-term investors **[5]**. Like all moving averages, this technical indicator is used to produce buy and sell signals based on crossovers and divergences from the historical average **[3]**. # 6. Learnings & Conclusions

The Polygon.io Stock Options API provides REST endpoints that let users query the latest market data from all US options exchanges. Users can find data on active and historical data. The free tier subscription of the API limits API requests to 5 API requests per minute, limits access to historical data (<1.5 years), and limits the number of base aggregates queried to create the aggregate results (Max = 50000). As shown in the "Variable Description" at the top of this notebook, data requests are set through several parameter options. One can get aggregate bars for an option contract over a given date range in custom time window sizes. For example, if timespan = ‘minute’ and multiplier = ‘5’ then 5-minute bars will be returned. Parameter options to sort results by timestamp are also available. For instance, the asc option will return results in ascending order (oldest at the top), and desc will return results in descending order (newest at the top). Other extended query parameters are also available (see References Section below for links to full list of query parameter options offered by the API). By default, the API returns all endpoints as a JSON response, which can be modified to a CSV response if so desired.

The most relevant query parameters are the stocksTicker, which specify a case-sensitive stock symbol (e.g., 'AAPL' for Apple Inc); the multiplier, which specify the size of the timespan multiplier; the timespan, which specify the size of the time window (e.g., by day, by week, etc., by second, by hour, etc.); and the start date and end date parameters, which specify the returned results lower and upper date ranges, respectively. In regards to the timestamps, the response returns data in UNIX timestamps (i.e., UTC/GMT time), and in miliseconds, we needed to write code lines to parse such data into a local timezone and YYYY-MM-DD format.

The JSON response returns valid data for the Stock Market trading periods, i.e., no endpoints with timespans outside the Open and Close periods are returned. Additionally, only endpoints with timespans during a stock's trading period (if not traded daily) are returned. These two features of the API response prevented the need of having to impute a variable's value for the analyses conducted herein.

The most relevant variables returned in the JSON response are found in the 'results', which are shown and described in the Variable Description at the top of this notebook. The variables contained in the 'results' provide the necessary information for traders and investors to understand the status of their stocks assets and the equity they hold. These variables are critical in many types of analysis directed at maximizing profits and minimalizing losses, for example, analyses for building stock portfolios, for devising a stock buying/selling strategy, etc. Herein, such variables were used for the computation of performance metrics and "indicators" relevant to a trader's buy or sell decision.

There are several indicators that can be used by investors or traders to make financially smart buy or sell decisions **[1]**. Some indicators are more specialized for particular types of investments (e.g., options or Cryptocurrency)  **[2]** **[3]**, and some are completely spurious in nature **[4]** **[5]**. They all have pros and cons, therefore, investors and traders typically use and compare two or more indicators for making holding or exiting decisions. There are also different approaches for using or testing a stock's performance with such indicators. For example, backtesting allows users to test their investment ideas or trading systems against historical data to evaluate their performance under various market conditions, while Forward testing, also known as paper trading, provides traders a mechanism for the simulation of actual trading and involves following the system's logic in a live market. It is also called paper trading since all trades are executed on paper only; that is, trade entries and exits are documented along with any profit or loss for the system, but no real trades are executed **[6]** In this notebook we use backtesting for three commonly employed trading indicators, namely, EMA, SMA, and RSI.
